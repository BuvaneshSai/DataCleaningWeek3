```python
import pandas as pd

# Read the TSV file into a DataFrame
df = pd.read_csv('/content/chipotle.tsv', sep='\t')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Save the DataFrame to a CSV file
df.to_csv('/content/chipotle.csv', index=False)

# Print a message indicating the conversion is complete
print("CSV file successfully created.")
```

```python
# Check for missing values in each column
missing_values = df.isna().sum()

# Print the missing values count for each column
print("Missing Values:")
print(missing_values)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
chipotle_df = pd.read_csv("/content/chipotle.csv")

# Replace missing values in the "Choice Description" column with "undetermined"
chipotle_df["choice_description"].fillna("undetermined", inplace=True)

# Display the DataFrame to verify the changes
print(chipotle_df.head())
```

```python
# Check for missing values in the DataFrame
missing_values = chipotle_df.isnull().sum()

# Display the count of missing values for each column
print(missing_values)
```

```python
# Verify the data types of each column
data_types = chipotle_df.dtypes

# Display the data types
print(data_types)
```

```python
print(chipotle_df['item_price'].dtype)
```

```python
chipotle_df['item_price'] = chipotle_df['item_price'].apply(lambda x: float(x.replace('$', '')) if isinstance(x, str) else x)
```

```python
chipotle_df.to_csv('/content/chipotle_cleaned.csv', index=False)
```
